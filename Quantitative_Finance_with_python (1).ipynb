Simple Rate of **Return**

Rate of return % = [(Current Value – Initial Value) / Initial Value] x 100

```python
PG["Simple_Return"] = (PG['Close']/PG['Close'].shift(1))-1
PG
```

```python
#alternatively
PG['returns'] = PG['Close'].pct_change(1)
PG
```

```python
PG.columns
```

```python
PG.drop(columns=['Open','High','Low','Volume',"Dividends",'Stock Splits'],inplace=True)
PG
```

```python
PG['Simple_Return'].plot(figsize=(12,8))
plt.show()
```

**Average daily return**

```python
avg_returns_d = PG['Simple_Return'].mean()
avg_returns_d
```

**Average annual returns¶**

```python
##very very less hence investors interested in annual returns

avg_returns_a = PG['Simple_Return'].mean()*250
avg_returns_a
```

```python
0.20795175008706676*100
```

```python
print (str(round(avg_returns_a,3) *100) + '%')
```

**Log returns**

#log returns are useful for single stock 
#simple returns are useful for multiple stocks

#formula of log returns = ln(p1/p0)= ln p1-ln p0

```python
PG['Log_Returns'] = np.log((PG['Close'])/PG['Close'].shift(1))
PG
```

```python
PG['Log_Returns'].plot(figsize=(12,8))
plt.show()
```

```python
log_returns_d = PG['Log_Returns'].mean()#avg daily log returns
log_returns_d
```

```python
log_returns_a = PG["Log_Returns"].mean()*250#average annual log returns
log_returns_a
```

```python
round(log_returns_a,5)*100
```

```python
print (str(round(log_returns_a, 3)*100) + '%')\
```

# Moving Average of three stocks

```python
#Let's plot some MA (Moving Averages). 
#Plot out the MA50 and MA200 for GM.

# Code here
gm['MA50'] = gm['Open'].rolling(50).mean()
gm['MA200'] = gm['Open'].rolling(200).mean()
gm[['Open','MA50','MA200']].plot(label='gm',figsize=(16,8))
```

# Correlogram of three stocks

```python
from pandas.plotting import scatter_matrix
car_comp = pd.concat([tesla['Open'],gm['Open'],ford['Open']],axis=1)
car_comp.columns = ['Tesla Open','GM Open','Ford Open']
# You can use a semi-colon to remove the axes print outs
scatter_matrix(car_comp,figsize=(8,8),alpha=0.2,hist_kwds={'bins':50});
```

```python
import seaborn as sns
sns.pairplot(car_comp)
```

```python
sns.heatmap(car_comp.corr(), annot=True)
```

```python
#understanding cummulative returns and daily returns
```

```python
#follwing will be code for cumulative daily returns
#df[daily_cumulative_return]=( 1 + df[pct_daily_return] ).cumprod()

ford.head()
```

```python
#Create a cumulative daily return .

ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
ford.head()
```

```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
tesla.head()
```

```python
gm['Cumulative Return'] = (1 + gm['returns']).cumprod()
gm.head()
```

```python
#Plot the Cumulative Return columns against the time series index.
tesla['Cumulative Return'].plot(label='Tesla',figsize=(16,8),title='Cumulative Return')
ford['Cumulative Return'].plot(label='Ford')
gm['Cumulative Return'].plot(label='GM')
plt.legend()
```
